```python
from IPython.display import display
import pandas as pd
```

```python
Global_df=pd.read_excel("Post-Marketing_Reports_reproductive_system.xlsx")
Global_df=pd.DataFrame(Global_df)
```

```python
def assemble(obj):
    word=""
    words=[]
    wordsList=[]
    rem=lambda x:x.strip(", () \n")
    for letter in obj:
        word=word+letter
        if letter==",":
            words.append(rem(word))
            word=""
        wordsList.append(words)
    return words
def getColumnObj(column=Global_df):
    letterInObj=[]
    letterInObjList=[]
    #your rows range -1 because i dont remember the argument for dataframe size
    for i in range(1739):
        #bj=df.iloc[i,[0,1,2,3,4,5]]
        for word in Global_df.loc[i,"Adverse Events"]:
            letterInObj.append(word)
        letterInObjList.append(assemble(letterInObj))
        letterInObj.clear()
    return letterInObjList


#print(getColumnObj())
```

```python
Adverse_eff_list=getColumnObj()
print(len(Adverse_eff_list))
```

```python
def isInt(s):
    try:
        int(s)
        return True
    except ValueError:
        return False
def genWordAndNumberList(list_of_words=Adverse_eff_list):
    events=[]
    words=[]
    contador_lista=0
    contador_palavra=0
    contador_digito=0
    # Adverse_eff_list is a list with of lists of words
    for lista in list_of_words:
        contador_lista+=1
        #print("Contador da lista %d"%contador_lista, "Contador da digito %d"%contador_digito, "Contador da palavra %d"%contador_palavra)
        for adverse_effect in lista:
            contador_palavra+=1
            num=""
            chk=""
            #print("Contador da palavra %d"%contador_palavra)
            for digit in adverse_effect:
                contador_digito+=1
                if isInt(digit):
                    num=num+digit
                    
                else:
                    chk=chk+digit
                #print("Contador da digito %d"%contador_digito)
            words.append([chk.strip("( "), contador_lista-1])
            events.append([num, contador_lista-1])
        contador_palavra=0


    return words,events
    # words=[word for word in assemble(genColumnObj()) if isInt(word)!=True]

effects,events=genWordAndNumberList()
print(events[0], len(events))
```
